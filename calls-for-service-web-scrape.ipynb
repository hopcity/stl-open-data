```python
import pandas
import time
import datetime
import os

today = datetime.datetime.now()

today_file_name = 'data/STL-calls-for-service_' + str(today).split(' ')[0] + '.csv'
print('File for today: ' + today_file_name)

if os.path.exists(today_file_name):
    print('Reading CSV from today')
    cfs = pandas.read_csv(today_file_name,index_col=False)
else:
    print('Creating First CSV for today')
    cfs = pandas.DataFrame(columns=['date','report_code','location','reported_issue'])

cfs_len = len(cfs)
print('Calls-for-service before now: ' + str(cfs_len))

new_cfs = pandas.read_html('http://slmpd.org/cfs.aspx')[0]
new_cfs.columns = ['date','report_code','location','reported_issue']

cfs = cfs.append(new_cfs).drop_duplicates()
cfs.reset_index(inplace=True,drop=True)

# Today only
cfs.loc[:,'date'] = pandas.to_datetime(cfs.date)
cfs = cfs[cfs.date.apply(lambda x: (x.year==today.year) & (x.month==today.month) & (x.day==today.day))]

cfs.reset_index(inplace=True,drop=True)
cfs.sort_values('date',inplace=True)
cfs.reset_index(inplace=True,drop=True)

cfs.to_csv(today_file_name,index=False)
print('New calls-for-service since last run: ' + str(len(cfs) - cfs_len))
print('All calls-for-service so far today: ' + str(len(cfs)))
```
